#### Data Cleaning Function

```python
def clean_data(df):
    '''
    Function to clean DataFrame. Column names are simplified, null and unnecessary values are imputed, and categorical columns are encoded.
    
    
    Return: Cleaned DataFrame, ready for analysis - final encoding still to be applied.
    
    '''
    
    # Dropping the unnecessary columns
    df.drop(['Unnamed: 0', 'id'], axis = 1, inplace = True)
    
    # Lower casing all column names
    df.columns = df.columns.str.lower()

    # Replacing space characters with underscores
    special_chars = "/ -" 
    for special_char in special_chars:
        df.columns = [col.replace(special_char, '_') for col in df.columns]
    
    # Simplifying answers, will also help when one-hot encoding categorical columns
    cat_cols = ['gender', 'customer_type', 'class', 'type_of_travel', 'satisfaction']

    for column in cat_cols:
        df[column] = df[column].str.lower() 
        
    # There are null values in the 'Arrival Delay Time' column, since we can't safely presume that these values equate to no delay, we'll fill with the mean of the column
    df['arrival_delay_in_minutes'].fillna(df['arrival_delay_in_minutes'].mean(), inplace = True)
    df.round({'arrival_delay_in_minutes' : 1})
    
    # Dropping '0' value rows from DataFrame, they don't serve any purpose, and they will help us simplify our data
    rank_list = ["inflight_wifi_service", "departure_arrival_time_convenient", "ease_of_online_booking", "gate_location",
                "food_and_drink", "online_boarding", "seat_comfort", "inflight_entertainment", "on_board_service",
                "leg_room_service", "baggage_handling", "checkin_service", "inflight_service", "cleanliness"]

    for col in rank_list:
        df.drop(df.loc[df[col]==0].index, inplace=True)

    # Converting 'satisfaction' column to a binary column
    df['satisfaction'].replace({'neutral or dissatisfied': 0, 'satisfied': 1},inplace = True)
    
    cleaned_df = df
    
    return cleaned_df
```

## III: Data Preparation

#### Encoding of Categorical Columns

Certain columns have string values while the rest are binary or numeric, encoding them will make these columns more easily ingestible by our models (and some models require all datatypes to be non string, non boolean).

```python
def encode_data(df):
    '''
    Encodes remaining categorical variables of data frame to be ready for model ingestion
    
    Inputs:
        customer_type (str): If the customer is a frequent flyer with the airline or not
        type_of_travel (str): Purpose of the trip taken
        gender (str): Gender of passenger
        class (str): Travel cabin in which the passenger flew
    
    return: Encoded Column Values
    '''
    
    df.customer_type.replace({'loyal customer': 1, 'disloyal customer': 0}, inplace = True)
    df.type_of_travel.replace({'personal travel': 1, 'business travel': 0}, inplace = True)
    df.gender.replace({'male': 1, 'female' : 0}, inplace = True)
    df['class'].replace({'eco': 'economy', 'eco plus': 'premium economy'}, inplace = True)  
    
    encoded_df = pd.get_dummies(df, columns = ['class'])
    
    return encoded_df
```

#### Assembling model architecture

```python
# Features as selected from feature importance
features = ['type_of_travel', 'flight_distance', 'inflight_wifi_service',
       'food_and_drink', 'online_boarding', 'seat_comfort',
       'inflight_entertainment', 'on_board_service', 'leg_room_service',
       'baggage_handling', 'checkin_service', 'inflight_service',
       'cleanliness', 'class_business', 'class_economy']

# Specifying target variable
target = ['satisfaction']

# Splitting into train and test
X_train = train_encoded[features].to_numpy()
X_test = train_encoded[target]
y_train = test_encoded[features].to_numpy()
y_test = test_encoded[target]
```

#### Model Activation Function

```python
# Time scores and metrics imports
import time
from sklearn.metrics import accuracy_score, roc_auc_score, plot_confusion_matrix, plot_roc_curve, precision_score, recall_score
```

```python
# Model activation and result plot function
def run_model(model, X_train, X_test, y_train, y_test):
   
    '''
    Model activation function. Function takes in model as a parameter (with respective model parameters applied if necessary), runs it again the train and test datasets,
    and outputs model metrics, performance, confusion matrix, and ROC AUC curve.
    
    Inputs: 
        model,  X_train, y_train, X_test, y_test
    
    Output: 
        Model Results, Confusion Matrix Plot, ROC_AUC Curve Plot
    '''
    
    t0 = time.time()

    model.fit(X_train,  y_train)
    y_pred = model.predict(X_test)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred) 
    time_taken = time.time() - t0
    memory_used = int(getrusage(RUSAGE_SELF).ru_maxrss / 1024)
    
    print("Accuracy = {}".format(accuracy))
    print("Precision = {}".format(precision))
    print("Recall = {}".format(recall))
    print("ROC Area under Curve = {}".format(roc_auc))
    print("Time taken = {} seconds".format(time_taken))
    print("Memory consumed = {} MB".format(memory_used))
    
    plot_confusion_matrix(model, X_test, y_test,cmap=plt.cm.Blues, normalize = 'all')
    plot_roc_curve(model, X_test, y_test)                     
    
    return model, accuracy, precision, recall, roc_auc, time_taken, memory_used   
```

#### Build #1 : AdaBoostClassifier Algorithm

```python
# Adaptive Booster Algorithm Assembly

# Parameters
ada_params = {'n_estimators': 1000,
              'random_state': 12345
                }

model_ada = AdaBoostClassifier(**ada_params)

# Saving output metrics
model_ada, accuracy_ada, precision_ada, recall_ada, roc_auc_ada, tt_ada, mu_ada = run_model(model_ada, X_train, y_train, X_test, y_test)
```

#### Build #2 : Logistic Regression

```python
# Logistic Regression Model Assembly

# Parameters (We won't be using any for this pipeline)
model_lr = LogisticRegression()
model_lr, accuracy_lr, precision_lr, recall_lr, roc_auc_lr, tt_lr, mu_lr = run_model(model_lr, X_train, y_train, X_test, y_test)
```

#### Build #3 : Random Forest Classifier Algorithm

```python
# Random Forest Classifier Assembly

# Parameters
model_rfc = RandomForestClassifier(n_estimators = 300)

# Saving Output Metrics
model_rfc, accuracy_rfc, precision_rfc, recall_rfc, roc_auc_rfc, tt_rfc, mu_rfc = run_model(model_rfc, X_train, y_train, X_test, y_test)
```

#### Build #4 : Multinomial Naive Bayes Algorithm

```python
# Multinomial Naive Bayes Model Assembly

# Parameters (We won't be using any for this pipeline)
model_mnb = MultinomialNB()

# Saving Output Metrics
model_mnb, accuracy_mnb, precision_mnb, recall_mnb, roc_auc_mnb, tt_mnb, mu_mnb = run_model(model_mnb, X_train, y_train, X_test, y_test)
```
